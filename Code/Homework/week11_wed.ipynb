```python
import pandas as pd 
data = pd.read_csv("test5-t.csv")
data.isnull().sum()
```

```python
# 删除有缺省值的行
data1=data.dropna()
data1
```

```python
# 删除有缺省值的列
data2=data.dropna(axis=1)
data2
```

```python
# 用0填充缺省值
data3=data.fillna(0)
data3
```

```python
# 用上一个数据填充缺省值
data4=data.fillna(method="pad")
data4
```

```python
# 用平均值填充缺省值
data4=data.fillna(data.mean())
data4
```
